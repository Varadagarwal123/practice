```python
import admin as ad
import student as st
import report as re
import read_write as rw
print('Welcome to Attendance Management System')
print('\n')
admin = {('varad', '123')}
database = rw.read_database_from_file()  

while True:
    print('1. Admin Login'
          '\n2. Student Login'
          '\n3. Exit')
    choice = int(input("Enter your choice: "))

    if choice == 1:
        username = input("Enter the username: ")
        password = input("Enter the password: ")

        if (username, password) in admin:
            print("Login Successful")
            while True:
                print("\n1. Register a new student"
                      "\n2. View the student database"
                      "\n3. Delete a student record"
                      "\n4. View Attendance of a student"
                      "\n5. Mark the attendance"
                      "\n6. Generate a daily report"
                      "\n7. Generate a weekly report"
                      "\n8. Generate a monthly report"
                      "\n9. Logout")
                option = int(input("Enter the option: "))

                if option == 1:
                    ad.register_student(database)
                elif option == 2:
                    ad.view_student_database(database)
                elif option == 3:
                    ad.delete_student_record(database)
                elif option == 4:
                    ad.view_student_attendance(database)
                elif option == 5:
                    ad.mark_attendance(database)
                elif option == 6:
                    re.generate_daily_report(database)
                elif option == 7:
                    re.generate_weekly_report(database)
                elif option == 8:
                    re.generate_monthly_report(database)
                elif option == 9:
                    rw.save_database_to_file(database)  
                    print("Logged out successfully!")
                    break
                else:
                    print("Enter a valid option.")
        else:
            print("Invalid credentials. Please try again.")

    elif choice == 2:
        if not database:  
            print("No students are registered. Please register as an admin first.")
        else:
            roll_number = input("Enter your roll number: ")
            password = input("Enter your password: ")

            if roll_number in database and database[roll_number]['password'] == password:
                print("Login Successful")
                while True:
                    print("\n1. View my attendance"
                          "\n2. Generate a daily report"
                          "\n3. Generate a weekly report"
                          "\n4. Generate a monthly report"
                          "\n5. Logout")
                    option = int(input("Enter the option: "))

                    if option == 1:
                        st.view_my_attendance(database, roll_number) 
                    elif option == 2:
                        ad.generate_daily_report(database)
                    elif option == 3:
                        ad.generate_weekly_report(database)
                    elif option == 4:
                        ad.generate_monthly_report(database)
                    elif option == 5:
                        print("Logged out successfully!")
                        break
                    else:
                        print("Enter a valid option.")
            else:
                print("Invalid credentials. Please try again.")

    elif choice == 3:
        rw.save_database_to_file(database) 
        print('Thank you')
        break

    else:
        print("Enter a valid option.")
```
